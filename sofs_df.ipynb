```python
import re
import pandas as pd
from nltk.tokenize import sent_tokenize
import nltk
```

```python
with open("sofs_text.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()


start = raw_text.find("===== c01.htm =====")
end = raw_text.find("===== cop.htm =====")

diary_text = raw_text[start:end]

sections = re.split(r"===== (c\d+)\.htm =====", diary_text)

chapters = []

for i in range(1, len(sections), 2):
    chapter_id = sections[i]
    text = sections[i+1]

    # extract the diary date from the text
    match = re.search(r"([A-Z][a-z]+ \d{1,2}, 1847)", text)
    date = match.group(1) if match else None

    chapters.append({
        "chapter_id": chapter_id,
        "date": date,
        "text": text.strip()
    })

chapters_df = pd.DataFrame(chapters)
chapters_df
```

```python
sentences = []

for _, row in chapters_df.iterrows():
    sents = sent_tokenize(row["text"])
    for s in sents:
        sentences.append({
            "chapter_id": row["chapter_id"],
            "date": row["date"],
            "sentence": s
        })

sent_df = pd.DataFrame(sentences)
sent_df.head()
```

```python
sample_df = sent_df.sample(frac=1, random_state=42).reset_index(drop=True)
sample_df.to_csv("sofs_sentences_samples.csv", index=False)
```
